## Types of Machine Learning Algorithms

There are three main categories of machine learning problems: **supervised learning**, **unsupervised learning**, and **reinforcement learning**. 


## Supervised Learning

In a supervised learning task, the goal is to generate a model that can predict an output based on one or many input values. 

* The output is typically called a **label**, **target value**, or **response variable**.
* The inputs are typically called **features** or **predictors**. 

The algorithm "learns" the optimal model by studying a training set of data that consists of several **observations** (or **instances**, or **samples**), each of which contains values for both its features and its label. 

Supervised learning tasks can be further classified into regression and classification problems. 

* **Regression.** A regression task is one for which the target values are continuous, real numbers. 
* **Classification.** A classification task is one for which the target values are discrete classes.

## Objective Functions

A supervised learning algorithm uses the data it is provided to learn the optimal model for the data it was provided. It determines which model is optimal by trying to either minimize or maximize an **objective function** that is used to score the model. 

## Training a Regression Algorithm:  Sum of Squared Errors

For regression tasks, a common objective function is the **sum of squared errors**, or **SSE**. Assume that our training set has $n$ observations, whose labels are denoted by $y_1, y_2, ..., y_n$. Assume that we create a regression model, and based on the features of the observations in our training set, the model predicts outputs of $\hat{y}_1, \hat{y}_2, ..., \hat{y}_n$. For each observation, we can define the error of the prediction to be $\hat{e}_i = \hat{y}_i - y_i$. Having calculated the errors for each of our training instances, we can then calculate the SSE as follows:

$\hspace{10 mm} \large SSE = \sum \hat{e}_i^2 $

The goal of the algorithm is the find the model that produces the lowest SSE. Consider the following example, which illustrates the linear regression algorithm.

## Unsupervised Learning

In unsupervised learning, we work with data sets in which features are provided, but no labels. Not only are the labels "missing", but we typically do not start an unsupervised learning tasks with a preconception of what the possible labels might be. The goal of unsupervised learning is to identify structure within the data that might not be readily perceptible to humans.
