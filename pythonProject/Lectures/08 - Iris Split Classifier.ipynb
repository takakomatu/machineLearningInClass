```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import MLUtilities as mlu
```

```python
iris = pd.read_table(filepath_or_buffer='Datasets/iris.txt', sep='\t')
iris.head(n=10)
```

```python
plt.close()
g = sns.pairplot(iris, hue="Species") # since we can only deal with three classes such as [.1231, .3234], we only can say is this 
#setosa or not? it this versicolor or not?
plt.show()
```

```python
iris2 = iris.copy()
```

```python
isSetosa = iris2.Species == 'setosa' # isSetosa looks like [true, false ....]
isSetosa = isSetosa.astype(int) # now looks like [1, 0 ...]
iris2['isSetosa'] = isSetosa

isVirginica = iris2.Species == 'virginica'
isVirginica = isVirginica.astype(int)
iris2['isVirginica'] = isVirginica

isVersicolor = iris2.Species == 'versicolor'
isVersicolor = isVersicolor.astype(int)
iris2['isVersicolor'] = isVersicolor

iris2.head()
```

## Classify Setosa

```python
X = iris2.iloc[:,0:4]
y = iris2.iloc[:,5]

#Threshold is .8 which is the line on Patal.Length and Patal.Width on 3 cells above
split_set = mlu.SplitClassifier(X,y)
split_set.summary()
```

## Classify Virginica

```python
X = iris2.iloc[:,0:4]
y = iris2.iloc[:,6]

split_set = mlu.SplitClassifier(X,y)
split_set.summary()
```

## Classify Versicolor

```python
X = iris2.iloc[:,0:4]
y = iris2.iloc[:,7]

split_set = mlu.SplitClassifier(X,y)
split_set.summary()
```
